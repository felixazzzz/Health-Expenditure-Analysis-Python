```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better visualization
sns.set_style("whitegrid")

# Define data
data = {
    "Year": [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015],
    "Argentina": [2.3, 3.25, 1.45, 8.82, -1.68, 2.32, 3.49, 1.87, 2.87, 1.97],
    "Brazil": [3.11, -1.03, 2.96, 0.96, 1.94, -1.16, -2.56, 2.67, -1.05, -1.81],
    "Colombia": [0.37, -1.99, -1.83, 3.88, 0.24, 1.76, -4.95, -0.01, 0.78, 0.61],
    "Costa Rica": [3.37, 0.15, 2.43, 4.61, 1.93, -2.93, 0.69, 0.45, 0.03, 2.37],
    "Cuba": [0.86, 5.27, 0.96, 0.91, -1.11, -1.97, -2.57, 1.04, 3.02, 0.09],
    "Dominican Republic": [9.7, 17.38, 12.85, 5.21, 4.48, 5.84, 1.19, 3.24, -9.23, -2.24],
    "Ecuador": [5.95, 20.2, 1.9, 7.03, 10.93, 2.11, 7.42, 10.63, 8.8, -5.93],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Define key points to highlight
highlight_points = {
    "Dominican Republic": {"year": 2007, "value": 17.38},  # Highest increase
    "Argentina": {"year": 2009, "value": 8.82},  # Most noticeable increase
    "Brazil": {"year": 2012, "value": -2.56},  # Largest negative growth
    "Ecuador": {"year": 2015, "value": -5.93},  # Most significant decline
}

# Define marker styles for highlighted points
highlight_styles = {
    "Dominican Republic": {"marker": "+", "color": "red"},  # Highest increase
    "Argentina": {"marker": "+", "color": "gold"},  # Most noticeable increase
    "Brazil": {"marker": "x", "color": "red"},  # Largest negative growth
    "Ecuador": {"marker": "x", "color": "gold"},  # Most significant decline
}

# Create the figure
plt.figure(figsize=(12, 6))

# Plot lines for each country
for country in df.columns[1:]:  # Exclude "Year" column
    plt.plot(df["Year"], df[country], label=country, marker='o', linestyle='-', linewidth=2)

# Highlight key points with different markers
for country, point in highlight_points.items():
    style = highlight_styles[country]
    plt.scatter(point["year"], point["value"], color=style["color"], s=150, marker=style["marker"], linewidth=2, zorder=3)

# Chart customization
plt.xlabel("Year", fontsize=12)
plt.ylabel("Year-over-Year Increase (%)", fontsize=12)
plt.title("Year-over-Year Increase in Government Health Expenditure (2006-2015)", fontsize=14, fontweight='bold')
plt.axhline(y=0, color='black', linewidth=1, linestyle='--')  # Reference line

# Adjust legend position inside the chart to save PPT space
plt.legend(fontsize=9, title="Country", loc="upper right", frameon=True, edgecolor="black")

# Optimize layout
plt.xticks(df["Year"], rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Save the figure
file_path = "Task3_figure.png"
plt.savefig(file_path, dpi=1200, bbox_inches="tight")

# Show the chart
plt.show()
```
